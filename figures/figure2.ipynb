```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data
plot_dict = {'1 worker': [20.98, 41.70, 65.607, 90, 108.79],
             '2 workers': [10.76, 23.62, 32.56, 42.1, 62.104],
             '3 workers': [7.64, 13.69, 19.55, 25.44, 38.40],
             '4 workers': [12.96, 24.75, 32.75, 46.45, 65.56]}
x_vals = [100, 200, 300, 400, 500]
```

```python
markers = ['o', 's', 'v', '^']
colors = ['red', 'blue', 'green', 'purple']
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=None) 
for i, (k, y_vals) in enumerate(plot_dict.items()):
  marker, color = markers[i], colors[i]
  plt.plot(x_vals, y_vals, marker=marker, color=color, label=k)
plt.legend()

plt.xticks(x_vals, x_vals)

# titles
plt.title('Inference time by varying number of workers and videos')
plt.xlabel('Number of videos')
plt.ylabel('Inference time (minutes)')

# despline
axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)
plt.savefig("figure2.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()
```
